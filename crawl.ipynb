```python
import pandas as pd
import numpy as np
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
```

```python
def clean_text(text):
    # Basic cleaning (adjust as needed)
    text = ' '.join(text.split())
    return text

def split_into_chunks(text, chunk_size=200, overlap=50):
    words = text.split()
    return [' '.join(words[i:i + chunk_size]) for i in range(0, len(words), chunk_size - overlap)]
```

```python

def process_data(df):
    model = SentenceTransformer('all-MiniLM-L6-v2')
    processed_data = []

    for _, row in df.iterrows():
        url = row['url']
        content = clean_text(row['content'])
        chunks = split_into_chunks(content)
        
        for chunk in chunks:
            embedding = model.encode(chunk)
            processed_data.append({
                'url': url,
                'chunk': chunk,
                'embedding': embedding.tolist()
            })

    return processed_data

```

```python

def retrieve_relevant_chunks(query, processed_data, top_k=3):
    model = SentenceTransformer('all-MiniLM-L6-v2')
    query_embedding = model.encode(query)
    
    similarities = [cosine_similarity([query_embedding], [np.array(item['embedding'])])[0][0] for item in processed_data]
    top_indices = np.argsort(similarities)[-top_k:][::-1]
    
    return [processed_data[i] for i in top_indices]
```

```python
def generate_answer(query, relevant_chunks):
    tokenizer = AutoTokenizer.from_pretrained("google/flan-t5-base")
    model = AutoModelForSeq2SeqLM.from_pretrained("google/flan-t5-base")

    context = " ".join([chunk['chunk'] for chunk in relevant_chunks])
    input_text = f"Answer the following question based on the given context. Question: {query} Context: {context}"
    
    input_ids = tokenizer(input_text, return_tensors="pt", max_length=512, truncation=True).input_ids
    outputs = model.generate(input_ids, max_length=150, num_return_sequences=1, temperature=0.7)
    answer = tokenizer.decode(outputs[0], skip_special_tokens=True)

    return answer
```
